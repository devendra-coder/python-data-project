```python
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt

# Loading Data
dataset = load_dataset("lukebarousse/data_jobs")
df = dataset['train'].to_pandas()

# Data CleanUp
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python
 median_yearly_salary = df['salary_year_avg'].median()
```

```python
median_hourly_salary = df['salary_hour_avg'].median()
```

```python
# filling missing salary values with median
df_filled = df
df_filled['salary_year_avg'] = df_filled['salary_year_avg'].fillna(median_yearly_salary)
df_filled['salary_hour_avg'] = df_filled['salary_hour_avg'].fillna(median_hourly_salary)
```

```python
# droping duplicates
df_unique = df_filled

df_unique = df_unique.drop_duplicates()

print('Lenght of original df:    ', len(df_filled))
print('Length of drop duplicates df:    ', len(df_unique))
print('rows dropped:    ', len(df_filled) - len(df_unique))
```

```python
df_unique  = df_unique.drop_duplicates(subset = ['job_title', 'company_name'])

print('Lenght of original df:    ', len(df_filled))
print('Length of drop duplicates df:    ', len(df_unique))
print('rows dropped:    ', len(df_filled) - len(df_unique))
```
